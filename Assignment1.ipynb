```python
mean_age = df['Age'].mean()
mean_age
```

```python
df.isnull().sum()
```

```python
df1 = df.dropna()
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
df1.describe()
```

```python
data_series = pd.Series(df1['Age']);

# Calculate Q1 and Q3
q1 = data_series.quantile(0.25)
q3 = data_series.quantile(0.75)

# Calculate the IQR
iqr = q3 - q1
lb = q1-1.5*iqr
ub = q3+1.5*iqr
print(iqr,lb,ub)
```

```python
columns_to_plot = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
plt.figure(figsize=(10, 6))
plt.boxplot([df[col] for col in columns_to_plot], labels=columns_to_plot)
plt.title("Boxplot of Selected Columns")
plt.ylabel('Values')
plt.show()
```

```python
df_cleaned = df[(df['Age'] >= lb) & (df['Age'] <= ub)]
df_cleaned
```

```python
columns_to_plot = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

# Plot multiple boxplots
plt.figure(figsize=(12, 10))
plt.boxplot([df_cleaned[col] for col in columns_to_plot], labels=columns_to_plot)
plt.title("Boxplot of Selected Columns")
plt.ylabel('Values')
plt.show()
```

Checking class distribution

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df_cleaned, palette='coolwarm')
plt.title("Survival Count")
plt.show()
```

# Step3 : Data Visualization

Gender-wise survival rate

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', hue='Sex', data=df_cleaned, palette='coolwarm')
plt.title("Survival Count by Gender")
plt.show()
```
